```python
from itertools import *
def create_pattern_list():
    patterns = []
    values = ['0','1','2']
    for a in values:
        for b in values:
            for c in values:
                for d in values:
                    for e in values:
                        temp = [a,b,c,d,e]
                        patterns.append(temp)
    return patterns
```
